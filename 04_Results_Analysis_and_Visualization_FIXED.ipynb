## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from matplotlib.patches import Rectangle
import warnings
warnings.filterwarnings('ignore')

# Set publication style
sns.set_context("paper", font_scale=1.3)
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']

print("✓ Libraries imported successfully")
```

## Figure 1: Model Cascading Performance (Combined Exp 01 & 02)

```python
print("Creating Figure 1: Model Cascading Performance...\n")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Define colors
color_exp1 = '#2E86AB'
color_exp2 = '#A23B72'

# Panel A: Traditional ML Cascade (Exp 01)
ax1.plot(exp1_df['llm_rate'] * 100, exp1_df['accuracy'],
         marker='o', linewidth=2.5, markersize=8, color=color_exp1, label='Cascade')

# Add baseline lines
small_model_acc_exp1 = 0.8180  # From your results
llm_acc_exp1 = 1.0000  # Oracle
ax1.axhline(y=small_model_acc_exp1, color='#F18F01', linestyle='--',
           linewidth=2, label=f'TF-IDF+LogReg ({small_model_acc_exp1:.3f})')
ax1.axhline(y=llm_acc_exp1, color='#E63946', linestyle='--',
           linewidth=2, label=f'Oracle ({llm_acc_exp1:.3f})')

ax1.set_xlabel('LLM Invocation Rate (%)', fontweight='bold', fontsize=12)
ax1.set_ylabel('Accuracy', fontweight='bold', fontsize=12)
ax1.set_title('(a) Traditional ML Cascade\n20 Newsgroups Dataset',
             fontweight='bold', fontsize=13)
ax1.legend(fontsize=10, loc='lower right')
ax1.grid(True, alpha=0.3, linestyle=':')
ax1.set_ylim([0.80, 1.01])

# Panel B: Neural Cascade (Exp 02)
ax2.plot(exp2_df['llm_rate'] * 100, exp2_df['accuracy'],
         marker='s', linewidth=2.5, markersize=8, color=color_exp2, label='Cascade')

# Add baseline lines
small_model_acc_exp2 = 0.9145  # From your results
llm_acc_exp2 = 0.9270  # FLAN-T5
ax2.axhline(y=small_model_acc_exp2, color='#F18F01', linestyle='--',
           linewidth=2, label=f'DistilBERT ({small_model_acc_exp2:.3f})')
ax2.axhline(y=llm_acc_exp2, color='#E63946', linestyle='--',
           linewidth=2, label=f'FLAN-T5 ({llm_acc_exp2:.3f})')

ax2.set_xlabel('LLM Invocation Rate (%)', fontweight='bold', fontsize=12)
ax2.set_ylabel('Accuracy', fontweight='bold', fontsize=12)
ax2.set_title('(b) Neural Cascade\nAG News Dataset',
             fontweight='bold', fontsize=13)
ax2.legend(fontsize=10, loc='lower right')
ax2.grid(True, alpha=0.3, linestyle=':')
ax2.set_ylim([0.90, 0.93])

plt.tight_layout()
plt.savefig('figures/final/Figure1_Model_Cascading.png', dpi=300, bbox_inches='tight')
plt.savefig('figures/final/Figure1_Model_Cascading.pdf', bbox_inches='tight')
print("✓ Saved: figures/final/Figure1_Model_Cascading.png")
print("✓ Saved: figures/final/Figure1_Model_Cascading.pdf")
plt.show()

print("\n✓ Figure 1 complete")
```

## Figure 2: Token-Aware Batching Performance (Exp 03)

```python
print("Creating Figure 2: Token-Aware Batching Performance...\n")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Prepare data
strategies = ['FIFO', 'Token-64', 'Token-128', 'Token-256']
padding_waste = exp3_df['Padding Waste (%)'].values
p95_latencies = exp3_df['P95 Latency (ms)'].values
throughputs = exp3_df['Throughput (req/s)'].values

# Panel A: Padding Waste Comparison
colors = ['#E63946', '#2A9D8F', '#2A9D8F', '#2A9D8F']
bars = ax1.bar(strategies, padding_waste, color=colors, alpha=0.8, edgecolor='black', linewidth=1.5)

# Highlight optimal
bars[2].set_edgecolor('#F18F01')
bars[2].set_linewidth(3)

ax1.set_ylabel('Padding Waste (%)', fontweight='bold', fontsize=12)
ax1.set_title('(a) Padding Waste by Strategy', fontweight='bold', fontsize=13)
ax1.grid(True, alpha=0.3, linestyle=':', axis='y')

# Add value labels
for bar, value in zip(bars, padding_waste):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
            f'{value:.1f}%', ha='center', va='bottom', fontweight='bold', fontsize=11)

# Add reduction annotation
ax1.annotate('', xy=(0, padding_waste[0]), xytext=(2, padding_waste[2]),
            arrowprops=dict(arrowstyle='<->', color='red', lw=2))
reduction = (padding_waste[0] - padding_waste[2]) / padding_waste[0] * 100
ax1.text(1, (padding_waste[0] + padding_waste[2])/2,
        f'{reduction:.1f}%\nreduction',
        ha='center', fontweight='bold', color='red', fontsize=10,
        bbox=dict(boxstyle='round', facecolor='white', edgecolor='red', alpha=0.8))

# Panel B: Latency vs Padding Trade-off
scatter_colors = ['#E63946', '#4ECDC4', '#F18F01', '#95E1D3']
for i, (strategy, color) in enumerate(zip(strategies, scatter_colors)):
    ax2.scatter(padding_waste[i], p95_latencies[i], s=300, c=color,
               alpha=0.8, edgecolors='black', linewidth=2, zorder=3)

    # Add labels with offsets
    offsets = [(10, -15), (10, 10), (10, -15), (10, 10)]
    ax2.annotate(strategy, (padding_waste[i], p95_latencies[i]),
                xytext=offsets[i], textcoords='offset points',
                fontsize=10, fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='white',
                         edgecolor='black', alpha=0.8))

# Highlight optimal (Token-128)
ax2.scatter(padding_waste[2], p95_latencies[2], s=400,
           facecolors='none', edgecolors='#F18F01', linewidth=4, zorder=2)

ax2.set_xlabel('Padding Waste (%)', fontweight='bold', fontsize=12)
ax2.set_ylabel('P95 Latency (ms)', fontweight='bold', fontsize=12)
ax2.set_title('(b) Padding vs Latency Trade-off', fontweight='bold', fontsize=13)
ax2.grid(True, alpha=0.3, linestyle=':')

# Add "optimal zone" annotation
ax2.text(padding_waste[2] + 2, p95_latencies[2] + 30, 'Optimal\nBalance',
        fontsize=9, fontweight='bold', color='#F18F01', ha='left')

plt.tight_layout()
plt.savefig('figures/final/Figure2_Token_Aware_Batching.png', dpi=300, bbox_inches='tight')
plt.savefig('figures/final/Figure2_Token_Aware_Batching.pdf', bbox_inches='tight')
print("✓ Saved: figures/final/Figure2_Token_Aware_Batching.png")
print("✓ Saved: figures/final/Figure2_Token_Aware_Batching.pdf")
plt.show()

print("\n✓ Figure 2 complete")
```

## Figure 3: Comprehensive System Performance

```python
print("Creating Figure 3: Comprehensive System Performance...\n")

fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(2, 3, hspace=0.3, wspace=0.3)

# Panel A: Exp01 - Accuracy vs LLM Rate
ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(exp1_df['llm_rate'] * 100, exp1_df['accuracy'],
         marker='o', linewidth=2.5, markersize=6, color='#2E86AB')
ax1.axhline(y=0.8180, color='#F18F01', linestyle='--', linewidth=2, alpha=0.7)
ax1.axhline(y=1.0, color='#E63946', linestyle='--', linewidth=2, alpha=0.7)
ax1.set_xlabel('LLM Rate (%)', fontweight='bold')
ax1.set_ylabel('Accuracy', fontweight='bold')
ax1.set_title('(a) Exp01: Traditional ML Cascade', fontweight='bold', fontsize=11)
ax1.grid(True, alpha=0.3, linestyle=':')

# Panel B: Exp01 - F1 Score vs LLM Rate
ax2 = fig.add_subplot(gs[0, 1])
ax2.plot(exp1_df['llm_rate'] * 100, exp1_df['f1_macro'],
         marker='o', linewidth=2.5, markersize=6, color='#A23B72')
ax2.set_xlabel('LLM Rate (%)', fontweight='bold')
ax2.set_ylabel('Macro F1-Score', fontweight='bold')
ax2.set_title('(b) Exp01: F1-Score vs LLM Rate', fontweight='bold', fontsize=11)
ax2.grid(True, alpha=0.3, linestyle=':')

# Panel C: Exp02 - Accuracy vs LLM Rate
ax3 = fig.add_subplot(gs[0, 2])
ax3.plot(exp2_df['llm_rate'] * 100, exp2_df['accuracy'],
         marker='s', linewidth=2.5, markersize=6, color='#2E86AB')
ax3.axhline(y=0.9145, color='#F18F01', linestyle='--', linewidth=2, alpha=0.7)
ax3.axhline(y=0.9270, color='#E63946', linestyle='--', linewidth=2, alpha=0.7)
ax3.set_xlabel('LLM Rate (%)', fontweight='bold')
ax3.set_ylabel('Accuracy', fontweight='bold')
ax3.set_title('(c) Exp02: Neural Cascade', fontweight='bold', fontsize=11)
ax3.grid(True, alpha=0.3, linestyle=':')

# Panel D: Exp03 - Padding Waste
ax4 = fig.add_subplot(gs[1, 0])
colors_d = ['#E63946', '#4ECDC4', '#F18F01', '#95E1D3']
bars_d = ax4.bar(strategies, padding_waste, color=colors_d, alpha=0.8, edgecolor='black')
ax4.set_ylabel('Padding Waste (%)', fontweight='bold')
ax4.set_title('(d) Exp03: Padding Waste', fontweight='bold', fontsize=11)
ax4.grid(True, alpha=0.3, linestyle=':', axis='y')
for bar, value in zip(bars_d, padding_waste):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height,
            f'{value:.1f}%', ha='center', va='bottom', fontweight='bold', fontsize=9)

# Panel E: Exp03 - Throughput
ax5 = fig.add_subplot(gs[1, 1])
bars_e = ax5.bar(strategies, throughputs, color=colors_d, alpha=0.8, edgecolor='black')
ax5.set_ylabel('Throughput (req/s)', fontweight='bold')
ax5.set_title('(e) Exp03: System Throughput', fontweight='bold', fontsize=11)
ax5.grid(True, alpha=0.3, linestyle=':', axis='y')
for bar, value in zip(bars_e, throughputs):
    height = bar.get_height()
    ax5.text(bar.get_x() + bar.get_width()/2., height,
            f'{value:.1f}', ha='center', va='bottom', fontweight='bold', fontsize=9)

# Panel F: Combined Cost Savings
ax6 = fig.add_subplot(gs[1, 2])

# Calculate combined savings
exp1_optimal_llm_rate = 0.49  # 51% reduction at θ=0.85
exp2_optimal_llm_rate = 0.063  # 93.7% reduction at θ=0.90
exp3_padding_reduction = 0.627  # 62.7% reduction with width=128

# Combined savings
exp1_combined = (1 - exp1_optimal_llm_rate) + (exp1_optimal_llm_rate * exp3_padding_reduction)
exp2_combined = (1 - exp2_optimal_llm_rate) + (exp2_optimal_llm_rate * exp3_padding_reduction)

savings_data = {
    'Component': ['Cascade Only\n(Exp01)', 'Batching Only\n(Exp03)',
                  'Combined\n(Exp01+03)', 'Combined\n(Exp02+03)'],
    'Savings': [51.0, 62.7, exp1_combined * 100, exp2_combined * 100]
}

colors_f = ['#2E86AB', '#A23B72', '#F18F01', '#E63946']
bars_f = ax6.bar(savings_data['Component'], savings_data['Savings'],
                color=colors_f, alpha=0.8, edgecolor='black')
ax6.set_ylabel('Cost Reduction (%)', fontweight='bold')
ax6.set_title('(f) Combined Cost Savings', fontweight='bold', fontsize=11)
ax6.set_ylim([0, 100])
ax6.grid(True, alpha=0.3, linestyle=':', axis='y')

for bar, value in zip(bars_f, savings_data['Savings']):
    height = bar.get_height()
    ax6.text(bar.get_x() + bar.get_width()/2., height,
            f'{value:.1f}%', ha='center', va='bottom', fontweight='bold', fontsize=9)

plt.savefig('figures/final/Figure3_Comprehensive_Performance.png', dpi=300, bbox_inches='tight')
plt.savefig('figures/final/Figure3_Comprehensive_Performance.pdf', bbox_inches='tight')
print("✓ Saved: figures/final/Figure3_Comprehensive_Performance.png")
print("✓ Saved: figures/final/Figure3_Comprehensive_Performance.pdf")
plt.show()

print("\n✓ Figure 3 complete")
```

## Generate Summary Tables for Paper

```python
print("Creating summary tables...\n")

# Table I: Experiment 01 Results (20 Newsgroups)
table1_data = [
    {'Method': 'TF-IDF+LogReg Only', 'Accuracy': 0.8180, 'F1 Macro': 0.8086,
     'LLM Rate': '0%', 'Latency (ms)': 12},
    {'Method': 'Oracle Only', 'Accuracy': 1.0000, 'F1 Macro': 1.0000,
     'LLM Rate': '100%', 'Latency (ms)': 548},
]

# Add selected cascade configurations from your actual results
selected_thresholds_exp1 = [0.80, 0.85, 0.90]
for threshold in selected_thresholds_exp1:
    row = exp1_df[exp1_df['threshold'] == threshold].iloc[0]
    table1_data.append({
        'Method': f'Cascade (θ={threshold:.2f})',
        'Accuracy': row['accuracy'],
        'F1 Macro': row['f1_macro'],
        'LLM Rate': f"{row['llm_rate']:.1%}",
        'Latency (ms)': int(row['avg_latency_ms'])
    })

table1_df = pd.DataFrame(table1_data)

print("="*80)
print("TABLE I: Traditional ML Cascade Results (20 Newsgroups)")
print("="*80)
print(table1_df.to_string(index=False))
print("="*80 + "\n")

# Save LaTeX
with open('tables/Table_I_Exp01.tex', 'w') as f:
    f.write(table1_df.to_latex(index=False, float_format="%.4f"))
print("✓ Saved: tables/Table_I_Exp01.tex\n")

# Table II: Experiment 02 Results (AG News)
table2_data = [
    {'Method': 'DistilBERT Only', 'Accuracy': 0.9145, 'F1 Macro': 0.9153,
     'LLM Rate': '0%', 'Latency (ms)': 7},
    {'Method': 'FLAN-T5 Only', 'Accuracy': 0.9270, 'F1 Macro': 0.9241,
     'LLM Rate': '100%', 'Latency (ms)': 672},
]

# Add selected cascade configurations
selected_thresholds_exp2 = [0.70, 0.75, 0.80, 0.90]
for threshold in selected_thresholds_exp2:
    row = exp2_df[exp2_df['threshold'] == threshold].iloc[0]
    table2_data.append({
        'Method': f'Cascade (θ={threshold:.2f})',
        'Accuracy': row['accuracy'],
        'F1 Macro': row['f1_macro'],
        'LLM Rate': f"{row['llm_rate']:.1%}",
        'Latency (ms)': int(row['avg_latency_ms'])
    })

table2_df = pd.DataFrame(table2_data)

print("="*80)
print("TABLE II: Neural Cascade Results (AG News)")
print("="*80)
print(table2_df.to_string(index=False))
print("="*80 + "\n")

# Save LaTeX
with open('tables/Table_II_Exp02.tex', 'w') as f:
    f.write(table2_df.to_latex(index=False, float_format="%.4f"))
print("✓ Saved: tables/Table_II_Exp02.tex\n")

# Table III: Experiment 03 Results (Token-Aware Batching)
table3_df = exp3_df.copy()
table3_df['Bucket Width'] = table3_df['Bucket Width'].replace('—', 'N/A')
table3_df = table3_df[['Strategy', 'Bucket Width', 'Padding Waste (%)',
                       'P95 Latency (ms)', 'Throughput (req/s)']]
table3_df['P95 Latency (ms)'] = table3_df['P95 Latency (ms)'].astype(int)

print("="*80)
print("TABLE III: Token-Aware Batching Results")
print("="*80)
print(table3_df.to_string(index=False))
print("="*80 + "\n")

# Save LaTeX
with open('tables/Table_III_Exp03.tex', 'w') as f:
    f.write(table3_df.to_latex(index=False, float_format="%.1f"))
print("✓ Saved: tables/Table_III_Exp03.tex\n")

print("✓ All tables created")
```

## Generate Executive Summary

```python
print("\n" + "="*80)
print("EXECUTIVE SUMMARY - ALL EXPERIMENTS")
print("="*80)

# Experiment 01 Best Results
exp1_optimal_idx = exp1_df['accuracy'].idxmax()
exp1_optimal = exp1_df.iloc[exp1_optimal_idx]

print("\n1. EXPERIMENT 01: Traditional ML Cascade (20 Newsgroups)")
print("   Dataset: 20 Newsgroups (20 classes)")
print("   Small Model: TF-IDF + Logistic Regression (81.8% accuracy)")
print("   Large Model: Oracle (100% accuracy)")
print(f"\n   OPTIMAL CONFIGURATION (θ={exp1_optimal['threshold']:.2f}):")
print(f"   → Accuracy: {exp1_optimal['accuracy']:.2%}")
print(f"   → F1-Score: {exp1_optimal['f1_macro']:.2%}")
print(f"   → LLM Reduction: {(1-exp1_optimal['llm_rate']):.1%}")
print(f"   → Average Latency: {exp1_optimal['avg_latency_ms']:.0f}ms")
print(f"\n   KEY INSIGHT: Achieved 98.1% accuracy while reducing LLM calls by 51%")

# Experiment 02 Best Results
exp2_optimal_idx = exp2_df['accuracy'].idxmax()
exp2_optimal = exp2_df.iloc[exp2_optimal_idx]

print("\n2. EXPERIMENT 02: Neural Cascade (AG News)")
print("   Dataset: AG News (4 classes)")
print("   Small Model: DistilBERT (91.45% accuracy)")
print("   Large Model: FLAN-T5 (92.7% accuracy)")
print(f"\n   OPTIMAL CONFIGURATION (θ={exp2_optimal['threshold']:.2f}):")
print(f"   → Accuracy: {exp2_optimal['accuracy']:.2%}")
print(f"   → F1-Score: {exp2_optimal['f1_macro']:.2%}")
print(f"   → LLM Reduction: {(1-exp2_optimal['llm_rate']):.1%}")
print(f"   → Average Latency: {exp2_optimal['avg_latency_ms']:.0f}ms")
print(f"\n   KEY INSIGHT: Achieved 91.8% accuracy while reducing LLM calls by 93.7%")

# Experiment 03 Best Results
fifo_row = exp3_df[exp3_df['Strategy'] == 'FIFO'].iloc[0]
token128_row = exp3_df[exp3_df['Bucket Width'] == 128].iloc[0]

padding_reduction = (fifo_row['Padding Waste (%)'] - token128_row['Padding Waste (%)']) / fifo_row['Padding Waste (%)'] * 100
throughput_gain = (token128_row['Throughput (req/s)'] - fifo_row['Throughput (req/s)']) / fifo_row['Throughput (req/s)'] * 100

print("\n3. EXPERIMENT 03: Token-Aware Batching")
print("   Workload: 5,000 requests with realistic token distribution")
print(f"   Baseline (FIFO): {fifo_row['Padding Waste (%)']:.1f}% padding waste")
print(f"\n   OPTIMAL CONFIGURATION (bucket_width=128):")
print(f"   → Padding Waste: {token128_row['Padding Waste (%)']:.1f}%")
print(f"   → Padding Reduction: {padding_reduction:.1f}%")
print(f"   → P95 Latency: {token128_row['P95 Latency (ms)']:.0f}ms")
print(f"   → Throughput: {token128_row['Throughput (req/s)']:.1f} req/s")
print(f"   → Throughput Gain: {throughput_gain:+.1f}%")
print(f"\n   KEY INSIGHT: Reduced padding waste by 62.7% with 6% throughput improvement")

# Combined System Performance
print("\n" + "="*80)
print("COMBINED SYSTEM PERFORMANCE")
print("="*80)

# Calculate combined savings
exp1_llm_rate = exp1_optimal['llm_rate']
exp2_llm_rate = exp2_optimal['llm_rate']
batching_padding_reduction = padding_reduction / 100

exp1_combined_savings = (1 - exp1_llm_rate) + (exp1_llm_rate * batching_padding_reduction)
exp2_combined_savings = (1 - exp2_llm_rate) + (exp2_llm_rate * batching_padding_reduction)

print("\nTraditional ML System (Exp01 + Exp03):")
print(f"   → Cascade reduces LLM calls by: {(1-exp1_llm_rate)*100:.1f}%")
print(f"   → For remaining {exp1_llm_rate*100:.1f}%, batching reduces padding by: {batching_padding_reduction*100:.1f}%")
print(f"   → TOTAL COMPUTATIONAL SAVINGS: {exp1_combined_savings*100:.1f}%")

print("\nNeural System (Exp02 + Exp03):")
print(f"   → Cascade reduces LLM calls by: {(1-exp2_llm_rate)*100:.1f}%")
print(f"   → For remaining {exp2_llm_rate*100:.1f}%, batching reduces padding by: {batching_padding_reduction*100:.1f}%")
print(f"   → TOTAL COMPUTATIONAL SAVINGS: {exp2_combined_savings*100:.1f}%")

print("\n" + "="*80)
print("PAPER-READY CLAIMS")
print("="*80)

print("\n1. Our traditional ML cascade achieves 98.1% accuracy while reducing")
print("   LLM invocations by 51%, demonstrating that simple classifiers can")
print("   effectively handle routine requests.")

print("\n2. Our neural cascade achieves 91.8% accuracy with 93.7% LLM reduction,")
print("   requiring expensive FLAN-T5 inference for only 6.3% of requests.")

print("\n3. Token-aware batching reduces padding waste by 62.7% compared to FIFO,")
print("   with bucket width 128 providing optimal balance between padding")
print("   efficiency and latency.")

print(f"\n4. Combined, our approach achieves {exp2_combined_savings*100:.1f}% computational cost")
print("   reduction while maintaining high accuracy, enabling sustainable")
print("   and efficient LLM serving at scale.")

print("\n" + "="*80)
print("ALL EXPERIMENTS COMPLETE")
print("="*80)

# Save summary
with open('experiment_results/EXECUTIVE_SUMMARY.txt', 'w') as f:
    f.write("="*80 + "\n")
    f.write("EXECUTIVE SUMMARY - ALL EXPERIMENTS\n")
    f.write("="*80 + "\n\n")
    f.write(f"Experiment 01: {exp1_optimal['accuracy']:.2%} accuracy, {(1-exp1_optimal['llm_rate']):.1%} LLM reduction\n")
    f.write(f"Experiment 02: {exp2_optimal['accuracy']:.2%} accuracy, {(1-exp2_optimal['llm_rate']):.1%} LLM reduction\n")
    f.write(f"Experiment 03: {padding_reduction:.1f}% padding reduction, {throughput_gain:+.1f}% throughput gain\n")
    f.write(f"\nCombined Savings: {exp2_combined_savings*100:.1f}%\n")

print("\n✓ Executive summary saved to experiment_results/EXECUTIVE_SUMMARY.txt")
```
