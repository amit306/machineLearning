```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, RobustScaler
from sklearn.metrics import (classification_report, confusion_matrix, roc_auc_score,
                           roc_curve, precision_recall_curve, f1_score, precision_score, recall_score)
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Input, Dense, Dropout, Embedding, Flatten,
                                   Concatenate, BatchNormalization, LeakyReLU)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l1_l2
from tensorflow.keras.utils import set_random_seed
import warnings
warnings.filterwarnings('ignore')

# Set seeds for reproducibility
set_random_seed(42)
tf.random.set_seed(42)
np.random.seed(42)

class FairnessAwareCreditModel:
    def __init__(self):
        self.encoders = {}
        self.scaler = None
        self.model = None
        self.history = None
        self.protected_attributes = ['Age', 'PersonalStatusSex', 'ForeignWorker']

    def load_and_preprocess_data(self):
        """Load and preprocess the German Credit dataset with fairness considerations"""
        print("ğŸ“Š Loading German Credit Dataset with Fairness Analysis...")

        # Load data
        url = "https://archive.ics.uci.edu/ml/machine-learning-databases/statlog/german/german.data"
        column_names = [
            "Status", "Duration", "CreditHistory", "Purpose", "CreditAmount", "Savings",
            "Employment", "InstallmentRate", "PersonalStatusSex", "Debtors", "ResidenceDuration",
            "Property", "Age", "OtherInstallmentPlans", "Housing", "ExistingCredits",
            "Job", "Liables", "Telephone", "ForeignWorker", "Target"
        ]

        df = pd.read_csv(url, sep=' ', header=None, names=column_names)
        df["Target"] = df["Target"].replace({1: 1, 2: 0})  # 1 = Good Credit, 0 = Bad Credit

        print(f"âœ… Dataset loaded: {df.shape[0]} samples, {df.shape[1]-1} features")

        # ğŸ” FAIRNESS: Create protected attribute indicators
        df = self._create_protected_attributes(df)

        # Analyze bias in original data
        self._analyze_dataset_bias(df)

        return df

    def _create_protected_attributes(self, df):
        """Create binary protected attributes for fairness analysis"""
        print("âš–ï¸ Creating protected attributes for fairness analysis...")

        # Age: Young (â‰¤25) vs Older (>25) - Young people might face discrimination
        df['age_binary'] = (df['Age'] <= 25).astype(int)
        print(f"ğŸ‘¥ Age groups: Young (â‰¤25): {df['age_binary'].sum()}, Older (>25): {(1-df['age_binary']).sum()}")

        # Gender: Extract from PersonalStatusSex (A91, A93, A94 = male; A92, A95 = female)
        male_codes = ['A91', 'A93', 'A94']
        df['gender_binary'] = df['PersonalStatusSex'].isin(male_codes).astype(int)  # 1=Male, 0=Female
        print(f"ğŸ‘¥ Gender: Male: {df['gender_binary'].sum()}, Female: {(1-df['gender_binary']).sum()}")

        # Foreign Worker: A201=Yes, A202=No
        df['foreign_binary'] = (df['ForeignWorker'] == 'A201').astype(int)  # 1=Foreign, 0=Domestic
        print(f"ğŸ‘¥ Foreign Worker: Foreign: {df['foreign_binary'].sum()}, Domestic: {(1-df['foreign_binary']).sum()}")

        return df

    def _analyze_dataset_bias(self, df):
        """Analyze potential bias in the original dataset"""
        print("\nğŸ” BIAS ANALYSIS IN ORIGINAL DATASET:")
        print("="*60)

        # Analyze bias for each protected attribute
        protected_attrs = {
            'age_binary': ('Young (â‰¤25)', 'Older (>25)'),
            'gender_binary': ('Female', 'Male'),
            'foreign_binary': ('Domestic', 'Foreign')
        }

        bias_results = {}

        for attr, (group0_name, group1_name) in protected_attrs.items():
            print(f"\nğŸ“Š Analysis for {attr.replace('_binary', '').title()}:")
            print("-" * 40)

            # Calculate approval rates by group
            group0_approval = df[df[attr] == 0]['Target'].mean()
            group1_approval = df[df[attr] == 1]['Target'].mean()

            # Calculate fairness metrics
            demographic_parity = abs(group1_approval - group0_approval)

            print(f"{group0_name} approval rate: {group0_approval:.3f}")
            print(f"{group1_name} approval rate: {group1_approval:.3f}")
            print(f"Demographic Parity Difference: {demographic_parity:.3f}")

            if demographic_parity > 0.1:
                print("âš ï¸  HIGH BIAS DETECTED!")
            elif demographic_parity > 0.05:
                print("âš ï¸  Moderate bias detected")
            else:
                print("âœ… Low bias")

            bias_results[attr] = {
                'group0_approval': group0_approval,
                'group1_approval': group1_approval,
                'demographic_parity': demographic_parity
            }

        return bias_results

    def calculate_fairness_metrics(self, y_true, y_pred, y_prob, protected_attr):
        """Calculate comprehensive fairness metrics"""
        metrics = {}

        # Demographic Parity Difference (Statistical Parity)
        group0_pred_rate = y_pred[protected_attr == 0].mean()
        group1_pred_rate = y_pred[protected_attr == 1].mean()
        metrics['demographic_parity'] = abs(group1_pred_rate - group0_pred_rate)

        # Equal Opportunity Difference (True Positive Rate equality)
        group0_tpr = y_pred[(protected_attr == 0) & (y_true == 1)].mean()
        group1_tpr = y_pred[(protected_attr == 1) & (y_true == 1)].mean()
        metrics['equal_opportunity'] = abs(group1_tpr - group0_tpr)

        # Equalized Odds (TPR and FPR equality)
        group0_fpr = y_pred[(protected_attr == 0) & (y_true == 0)].mean()
        group1_fpr = y_pred[(protected_attr == 1) & (y_true == 0)].mean()
        fpr_diff = abs(group1_fpr - group0_fpr)
        metrics['equalized_odds'] = max(metrics['equal_opportunity'], fpr_diff)

        # Predictive Rate Parity (Precision equality)
        group0_precision = y_true[(protected_attr == 0) & (y_pred == 1)].mean()
        group1_precision = y_true[(protected_attr == 1) & (y_pred == 1)].mean()
        metrics['predictive_rate_parity'] = abs(group1_precision - group0_precision)

        # Overall Accuracy Equality
        group0_accuracy = (y_pred[protected_attr == 0] == y_true[protected_attr == 0]).mean()
        group1_accuracy = (y_pred[protected_attr == 1] == y_true[protected_attr == 1]).mean()
        metrics['accuracy_equality'] = abs(group1_accuracy - group0_accuracy)

        return metrics

    def prepare_data_with_fairness(self, df):
        """Prepare data for training with fairness considerations"""
        print("ğŸ› ï¸ Preparing data with fairness considerations...")

        # Store protected attributes separately
        protected_data = df[['age_binary', 'gender_binary', 'foreign_binary']].copy()

        # Separate features and target (exclude protected attributes from features)
        feature_cols = [col for col in df.columns if col not in
                       ['Target', 'age_binary', 'gender_binary', 'foreign_binary']]
        X = df[feature_cols]
        y = df["Target"]

        # Identify column types
        categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
        numerical_cols = X.select_dtypes(include=[np.number]).columns.tolist()

        print(f"ğŸ“‹ Features used: {len(feature_cols)} (protected attributes excluded)")
        print(f"ğŸ“‹ Categorical features: {len(categorical_cols)}")
        print(f"ğŸ“‹ Numerical features: {len(numerical_cols)}")

        # Encode categorical variables
        for col in categorical_cols:
            le = LabelEncoder()
            X[col] = le.fit_transform(X[col].astype(str))
            self.encoders[col] = le

        # Split data while preserving protected attribute distribution
        X_train, X_test, y_train, y_test, prot_train, prot_test = train_test_split(
            X, y, protected_data, test_size=0.2, random_state=42, stratify=y
        )

        # Further split training data for validation
        X_train, X_val, y_train, y_val, prot_train, prot_val = train_test_split(
            X_train, y_train, prot_train, test_size=0.2, random_state=42, stratify=y_train
        )

        # Apply fairness-aware reweighting to training data
        X_train_reweighted, y_train_reweighted, sample_weights = self._apply_reweighting(
            X_train, y_train, prot_train['age_binary']
        )

        # Scale numerical features
        self.scaler = RobustScaler()
        X_train_num = self.scaler.fit_transform(X_train[numerical_cols])
        X_val_num = self.scaler.transform(X_val[numerical_cols])
        X_test_num = self.scaler.transform(X_test[numerical_cols])

        print(f"âœ… Data split - Train: {len(X_train)}, Val: {len(X_val)}, Test: {len(X_test)}")

        return {
            'X_train': X_train, 'X_val': X_val, 'X_test': X_test,
            'y_train': y_train, 'y_val': y_val, 'y_test': y_test,
            'X_train_num': X_train_num, 'X_val_num': X_val_num, 'X_test_num': X_test_num,
            'prot_train': prot_train, 'prot_val': prot_val, 'prot_test': prot_test,
            'sample_weights': sample_weights,
            'categorical_cols': categorical_cols, 'numerical_cols': numerical_cols
        }

    def _apply_reweighting(self, X, y, protected_attr):
        """Apply fairness-aware reweighting (preprocessing bias mitigation)"""
        print("âš–ï¸ Applying fairness-aware reweighting...")

        # Calculate reweighting factors for demographic parity
        sample_weights = np.ones(len(X))

        # Calculate group proportions
        for s in [0, 1]:  # protected attribute values
            for y_val in [0, 1]:  # target values
                # Find samples in this group
                mask = (protected_attr == s) & (y == y_val)
                n_group = mask.sum()

                if n_group > 0:
                    # Calculate desired weight for demographic parity
                    p_s = (protected_attr == s).mean()
                    p_y = (y == y_val).mean()
                    p_sy = mask.mean()

                    # Reweighting factor
                    weight = (p_s * p_y) / p_sy if p_sy > 0 else 1.0
                    sample_weights[mask] = weight

        # Normalize weights
        sample_weights = sample_weights / sample_weights.mean()

        print(f"âœ… Applied reweighting. Weight range: {sample_weights.min():.3f} - {sample_weights.max():.3f}")

        return X, y, sample_weights

    def build_fairness_aware_model(self, data_dict):
        """Build a fairness-aware neural network"""
        print("ğŸ—ï¸ Building fairness-aware neural network...")

        categorical_cols = data_dict['categorical_cols']
        numerical_cols = data_dict['numerical_cols']
        X_train = data_dict['X_train']

        # Embedding inputs for categorical features
        embed_inputs = []
        embed_layers = []

        for col in categorical_cols:
            input_layer = Input(shape=(1,), name=f'input_{col}')
            vocab_size = X_train[col].nunique()
            embed_dim = min(50, max(4, (vocab_size + 1) // 2))

            embedding = Embedding(
                input_dim=vocab_size,
                output_dim=embed_dim,
                embeddings_regularizer=l1_l2(l1=1e-5, l2=1e-4),
                name=f'embed_{col}'
            )(input_layer)
            embedding = Flatten()(embedding)
            embedding = Dropout(0.3)(embedding)  # Higher dropout for fairness

            embed_inputs.append(input_layer)
            embed_layers.append(embedding)

        # Numerical input with fairness-aware processing
        numeric_input = Input(shape=(len(numerical_cols),), name='numeric_input')
        numeric_layer = BatchNormalization()(numeric_input)
        numeric_layer = Dense(32, activation='relu',
                             kernel_regularizer=l1_l2(l1=1e-4, l2=1e-3))(numeric_layer)  # Stronger regularization
        numeric_layer = Dropout(0.4)(numeric_layer)

        embed_inputs.append(numeric_input)
        embed_layers.append(numeric_layer)

        # Combine all inputs
        if len(embed_layers) > 1:
            combined = Concatenate()(embed_layers)
        else:
            combined = embed_layers[0]

        # Fairness-aware deep network (simpler to reduce bias amplification)
        x = Dense(64, kernel_regularizer=l1_l2(l1=1e-4, l2=1e-3))(combined)
        x = BatchNormalization()(x)
        x = LeakyReLU(alpha=0.1)(x)
        x = Dropout(0.5)(x)  # High dropout for fairness

        x = Dense(32, kernel_regularizer=l1_l2(l1=1e-4, l2=1e-3))(x)
        x = BatchNormalization()(x)
        x = LeakyReLU(alpha=0.1)(x)
        x = Dropout(0.4)(x)

        # Output layer
        output = Dense(1, activation='sigmoid', name='output')(x)

        # Create model
        self.model = Model(inputs=embed_inputs, outputs=output, name='FairnessAwareCreditModel')

        # Compile with conservative learning rate for fairness
        optimizer = Adam(learning_rate=0.0005, beta_1=0.9, beta_2=0.999)
        self.model.compile(
            optimizer=optimizer,
            loss='binary_crossentropy',
            metrics=['accuracy']
        )

        print(f"âœ… Fairness-aware model built with {self.model.count_params():,} parameters")
        return self.model

    def train_with_fairness_constraints(self, data_dict, epochs=100):
        """Train the model with fairness constraints"""
        print("ğŸš€ Training fairness-aware model...")

        # Prepare inputs
        X_train_inputs = self._prepare_inputs(data_dict['X_train'], data_dict['X_train_num'],
                                            data_dict['categorical_cols'])
        X_val_inputs = self._prepare_inputs(data_dict['X_val'], data_dict['X_val_num'],
                                          data_dict['categorical_cols'])

        # Fairness-aware callbacks
        callbacks = [
            EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True, verbose=1),
            ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=15, min_lr=1e-7, verbose=1),
        ]

        # Train with sample weights (reweighting for fairness)
        self.history = self.model.fit(
            X_train_inputs, data_dict['y_train'],
            sample_weight=data_dict['sample_weights'],  # Fairness reweighting
            validation_data=(X_val_inputs, data_dict['y_val']),
            epochs=epochs,
            batch_size=64,  # Larger batch for stability
            callbacks=callbacks,
            verbose=1
        )

        print("âœ… Fairness-aware training completed!")
        return self.history

    def _prepare_inputs(self, X_df, X_num, categorical_cols):
        """Prepare inputs for the model"""
        inputs = []
        for col in categorical_cols:
            inputs.append(np.array(X_df[col]).reshape(-1, 1))
        inputs.append(X_num)
        return inputs

    def comprehensive_fairness_evaluation(self, data_dict):
        """Comprehensive evaluation including fairness metrics"""
        print("ğŸ“Š Comprehensive fairness evaluation...")

        # Prepare test inputs
        X_test_inputs = self._prepare_inputs(data_dict['X_test'], data_dict['X_test_num'],
                                           data_dict['categorical_cols'])

        # Predictions
        y_pred_prob = self.model.predict(X_test_inputs).flatten()
        y_pred_binary = (y_pred_prob >= 0.5).astype(int)

        y_test = data_dict['y_test']
        prot_test = data_dict['prot_test']

        # Performance metrics
        accuracy = np.mean(y_pred_binary == y_test)
        precision = precision_score(y_test, y_pred_binary)
        recall = recall_score(y_test, y_pred_binary)
        f1 = f1_score(y_test, y_pred_binary)
        auc_roc = roc_auc_score(y_test, y_pred_prob)

        print(f"\nğŸ¯ MODEL PERFORMANCE METRICS:")
        print(f"{'='*50}")
        print(f"Accuracy:  {accuracy:.4f}")
        print(f"Precision: {precision:.4f}")
        print(f"Recall:    {recall:.4f}")
        print(f"F1-Score:  {f1:.4f}")
        print(f"AUC-ROC:   {auc_roc:.4f}")

        # Fairness metrics for each protected attribute
        print(f"\nâš–ï¸ FAIRNESS METRICS:")
        print(f"{'='*70}")

        fairness_results = {}
        protected_attrs = {
            'age_binary': 'Age (Young vs Older)',
            'gender_binary': 'Gender (Female vs Male)',
            'foreign_binary': 'Foreign Worker (Domestic vs Foreign)'
        }

        for attr, desc in protected_attrs.items():
            print(f"\nğŸ“Š {desc}:")
            print("-" * 50)

            fairness_metrics = self.calculate_fairness_metrics(
                y_test.values, y_pred_binary, y_pred_prob, prot_test[attr].values
            )

            fairness_results[attr] = fairness_metrics

            for metric, value in fairness_metrics.items():
                status = "âœ… Fair" if value < 0.1 else ("âš ï¸ Moderate bias" if value < 0.2 else "âŒ High bias")
                print(f"{metric.replace('_', ' ').title():<25}: {value:.4f} {status}")

        return {
            'performance': {'accuracy': accuracy, 'precision': precision, 'recall': recall,
                           'f1_score': f1, 'auc_roc': auc_roc},
            'fairness': fairness_results,
            'predictions': {'y_pred_prob': y_pred_prob, 'y_pred_binary': y_pred_binary}
        }

    def plot_fairness_analysis(self, results, data_dict):
        """Plot comprehensive fairness analysis"""
        y_test = data_dict['y_test']
        prot_test = data_dict['prot_test']
        y_pred_prob = results['predictions']['y_pred_prob']
        y_pred_binary = results['predictions']['y_pred_binary']

        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle('Comprehensive Fairness Analysis', fontsize=16, fontweight='bold')

        protected_attrs = {
            'age_binary': ('Older (>25)', 'Young (â‰¤25)'),
            'gender_binary': ('Female', 'Male'),
            'foreign_binary': ('Domestic', 'Foreign')
        }

        for i, (attr, (group0_name, group1_name)) in enumerate(protected_attrs.items()):
            # Approval rates by group
            group0_approval = y_pred_binary[prot_test[attr] == 0].mean()
            group1_approval = y_pred_binary[prot_test[attr] == 1].mean()

            # Plot approval rates
            axes[0, i].bar([group0_name, group1_name], [group0_approval, group1_approval],
                          color=['lightblue', 'lightcoral'], alpha=0.7)
            axes[0, i].set_title(f'Approval Rates by {attr.replace("_binary", "").title()}')
            axes[0, i].set_ylabel('Approval Rate')
            axes[0, i].set_ylim(0, 1)

            # Add values on bars
            axes[0, i].text(0, group0_approval + 0.02, f'{group0_approval:.3f}',
                           ha='center', va='bottom', fontweight='bold')
            axes[0, i].text(1, group1_approval + 0.02, f'{group1_approval:.3f}',
                           ha='center', va='bottom', fontweight='bold')

            # Demographic parity difference
            dpd = abs(group1_approval - group0_approval)
            axes[0, i].text(0.5, 0.9, f'DPD: {dpd:.3f}', transform=axes[0, i].transAxes,
                           ha='center', va='center', bbox=dict(boxstyle="round,pad=0.3",
                           facecolor="yellow" if dpd > 0.1 else "lightgreen", alpha=0.7),
                           fontweight='bold')

            # Plot prediction score distributions
            group0_scores = y_pred_prob[prot_test[attr] == 0]
            group1_scores = y_pred_prob[prot_test[attr] == 1]

            axes[1, i].hist(group0_scores, bins=20, alpha=0.6, label=group0_name,
                           color='lightblue', density=True)
            axes[1, i].hist(group1_scores, bins=20, alpha=0.6, label=group1_name,
                           color='lightcoral', density=True)
            axes[1, i].axvline(x=0.5, color='black', linestyle='--', linewidth=2,
                              label='Threshold')
            axes[1, i].set_title(f'Score Distribution by {attr.replace("_binary", "").title()}')
            axes[1, i].set_xlabel('Prediction Score')
            axes[1, i].set_ylabel('Density')
            axes[1, i].legend()
            axes[1, i].grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

        # Fairness metrics summary plot
        self._plot_fairness_metrics_summary(results['fairness'])

    def _plot_fairness_metrics_summary(self, fairness_results):
        """Plot summary of all fairness metrics"""
        fig, ax = plt.subplots(figsize=(12, 8))

        metrics = ['demographic_parity', 'equal_opportunity', 'equalized_odds',
                  'predictive_rate_parity', 'accuracy_equality']
        attrs = ['age_binary', 'gender_binary', 'foreign_binary']
        attr_names = ['Age', 'Gender', 'Foreign Worker']

        x = np.arange(len(metrics))
        width = 0.25

        colors = ['skyblue', 'lightcoral', 'lightgreen']

        for i, (attr, name) in enumerate(zip(attrs, attr_names)):
            values = [fairness_results[attr][metric] for metric in metrics]
            bars = ax.bar(x + i*width, values, width, label=name, color=colors[i], alpha=0.8)

            # Add value labels on bars
            for bar, value in zip(bars, values):
                height = bar.get_height()
                ax.text(bar.get_x() + bar.get_width()/2., height + 0.005,
                       f'{value:.3f}', ha='center', va='bottom', fontsize=9)

        ax.set_xlabel('Fairness Metrics')
        ax.set_ylabel('Bias Level (Lower is Better)')
        ax.set_title('Fairness Metrics Summary Across Protected Attributes', fontweight='bold')
        ax.set_xticks(x + width)
        ax.set_xticklabels([m.replace('_', ' ').title() for m in metrics], rotation=45)
        ax.legend()
        ax.grid(True, alpha=0.3)

        # Add fairness threshold lines
        ax.axhline(y=0.1, color='orange', linestyle='--', alpha=0.7, label='Moderate Bias Threshold')
        ax.axhline(y=0.2, color='red', linestyle='--', alpha=0.7, label='High Bias Threshold')

        plt.tight_layout()
        plt.show()

def main():
    """Main execution function with fairness analysis"""
    print("âš–ï¸ FAIRNESS-AWARE GERMAN CREDIT RISK PREDICTION")
    print("="*70)

    # Initialize fairness-aware model
    credit_model = FairnessAwareCreditModel()

    # Load and preprocess data with fairness analysis
    df = credit_model.load_and_preprocess_data()
    data_dict = credit_model.prepare_data_with_fairness(df)

    # Build fairness-aware model
    model = credit_model.build_fairness_aware_model(data_dict)

    # Display model architecture
    print(f"\nğŸ—ï¸ FAIRNESS-AWARE MODEL ARCHITECTURE:")
    model.summary()

    # Train model with fairness constraints
    history = credit_model.train_with_fairness_constraints(data_dict, epochs=80)

    # Comprehensive evaluation including fairness
    results = credit_model.comprehensive_fairness_evaluation(data_dict)

    # Plot fairness analysis
    credit_model.plot_fairness_analysis(results, data_dict)

    print(f"\nğŸ‰ FAIRNESS-AWARE ANALYSIS COMPLETE!")
    print(f"âœ… Model trained with bias mitigation techniques")
    print(f"âœ… Comprehensive fairness evaluation completed")
    print(f"âœ… Multiple protected attributes analyzed")
    print(f"âœ… Reweighting applied for demographic parity")

# Execute the main function
if __name__ == "__main__":
    main()
```
